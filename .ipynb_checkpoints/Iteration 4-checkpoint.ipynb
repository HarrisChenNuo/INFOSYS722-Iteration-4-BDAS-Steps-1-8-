```python
# List of grouped files
mort_files = ['MORT_100.csv', 'MORT_200.csv', 'MORT_300.csv']
other_files = ['CHILDMORT_DEATHS_10TO14.csv', 'CHILDMORT_MORTALITY_10TO14.csv', 'CM_01.csv', 
               'CM_02.csv', 'CM_03.csv', 'CM_04.csv', 'MDG_0000000001.csv', 'MDG_0000000007.csv',
               'WHOSIS_000003.csv', 'WHOSIS_000014.csv', 'WHOSIS_000016.csv']
# Load and merge MORT files 
mort_dataframes = [spark.read.csv(f"{path_to_data}{file}", header=True, inferSchema=True) for file in mort_files]
combined_mort_data = mort_dataframes[0]
for df in mort_dataframes[1:]:
    combined_mort_data = combined_mort_data.unionByName(df)
# Load and merge other files 
other_dataframes = [spark.read.csv(f"{path_to_data}{file}", header=True, inferSchema=True) for file in other_files]
combined_other_data = other_dataframes[0]
for df in other_dataframes[1:]:
    combined_other_data = combined_other_data.unionByName(df)
```

```python
def find_outliers_extremes_spark(df):
    outlier_info = []
    # Get columns of numeric type 
    numeric_columns = [field.name for field in df.schema.fields if isinstance(field.dataType, (IntegerType, FloatType, DoubleType))]

    for column in numeric_columns:
        # Calculating Quartiles and IQR 
        quantiles = df.approxQuantile(column, [0.25, 0.75], 0.05)
        Q1 = quantiles[0]
        Q3 = quantiles[1]
        IQR = Q3 - Q1

        # Defining conditions for outliers and extremes 
        outliers_condition = ((col(column) < (Q1 - 1.5 * IQR)) | (col(column) > (Q3 + 1.5 * IQR)))
        extremes_condition = ((col(column) < (Q1 - 3 * IQR)) | (col(column) > (Q3 + 3 * IQR)))

        # Calculate the number of outliers and extremes 
        outliers_count = df.filter(outliers_condition).count()
        extremes_count = df.filter(extremes_condition).count()

        outlier_info.append((column, outliers_count, extremes_count))

    return outlier_info

# Application to the MORT dataset 
outliers_mort = find_outliers_extremes_spark(combined_mort_data)

# Apply to other datasets 
outliers_other = find_outliers_extremes_spark(combined_other_data)

# Print results 
print("Outliers and Extremes in MORT Data:")
for info in outliers_mort:
    print(f"Column: {info[0]}, Outliers: {info[1]}, Extremes: {info[2]}")

print("\nOutliers and Extremes in Other Data:")
for info in outliers_other:
    print(f"Column: {info[0]}, Outliers: {info[1]}, Extremes: {info[2]}")
```

```python
def calculate_missing_percentage(df):
    # Calculation of percentage of missing values 
    total_count = df.count()
    missing_percentage = df.select([(count(lit(1)) - count(c)).alias(c) for c in df.columns])
    missing_percentage = missing_percentage.select([((col(c) / total_count) * 100).alias(c) for c in df.columns])
    return missing_percentage

# Calculate the percentage of missing values for each column in the MORT dataset 
mort_missing_percentage = calculate_missing_percentage(combined_mort_data)
print("Percentage of missing values in MORT data:")
mort_missing_percentage.show()

# Calculate the percentage of missing values for each column in other datasets 
other_missing_percentage = calculate_missing_percentage(combined_other_data)
print("Percentage of missing values in other data:")
other_missing_percentage.show()
```

```python
def count_blanks(df):
    blank_counts = {}
    # Iterate over each column, checking for columns with a data type of string 
    for column in df.columns:
        if isinstance(df.schema[column].dataType, StringType):
            # Calculate the number of rows with empty strings or just blank characters in each column 
            count = df.filter((col(column) == '') | (col(column).rlike('^\s+$'))).count()
            blank_counts[column] = count
    return blank_counts

# Checking for empty strings in the MORT dataset 
mort_blanks = count_blanks(combined_mort_data)
print("Blank strings in MORT data:")
print(mort_blanks)

# Checking for empty strings in other datasets 
other_blanks = count_blanks(combined_other_data)
print("Blank strings in other data:")
print(other_blanks)
```

```python
# Only keep lines with 'DisaggregatingDimension1ValueCode' or 'DisaggregatingDimension2ValueCode' equal to 'DAYS0-27'. 
mort_300_filtered = mort_300_cleaned.filter(
    (col('DisaggregatingDimension1ValueCode') == 'DAYS0-27') | 
    (col('DisaggregatingDimension2ValueCode') == 'DAYS0-27')
)

# Display of filtered data 
mort_300_filtered.show(truncate=False)
```

```python
# Conversion function that rearranges DisaggregatingDimensions columns based on SpatialDimension values
mort_300_filtered = mort_300_filtered.withColumn(
    "DisaggregatingDimension1",
    when(
        (col("SpatialDimension") == 'MGHEREG') & (col("DisaggregatingDimension1") == 'SEX'),
        lit('AGEGROUP')
    ).otherwise(col("DisaggregatingDimension1"))
).withColumn(
    "DisaggregatingDimension1ValueCode",
    when(
        (col("SpatialDimension") == 'MGHEREG') & (col("DisaggregatingDimension1") == 'AGEGROUP'),
        lit('DAYS0-27')
    ).otherwise(col("DisaggregatingDimension1ValueCode"))
).withColumn(
    "DisaggregatingDimension2",
    when(
        (col("SpatialDimension") == 'MGHEREG') & (col("DisaggregatingDimension1") == 'AGEGROUP'),
        lit('CHILDCAUSE')
    ).otherwise(col("DisaggregatingDimension2"))
).withColumn(
    "DisaggregatingDimension2ValueCode",
    when(
        (col("SpatialDimension") == 'MGHEREG') & (col("DisaggregatingDimension2") == 'CHILDCAUSE'),
        col("DisaggregatingDimension3ValueCode")
    ).otherwise(col("DisaggregatingDimension2ValueCode"))
).withColumn(
    "DisaggregatingDimension3", lit(None).cast(StringType())
).withColumn(
    "DisaggregatingDimension3ValueCode", lit(None).cast(StringType())
)

# Display of converted data 
mort_300_filtered.show(truncate=False)
```

```python
# Reading the CHILDCAUSE.csv file 
child_cause = spark.read.csv(path_to_codes + 'CHILDCAUSE.csv', header=True, inferSchema=True)

# Keep only the Code and Title columns 
child_cause = child_cause.select('Code', 'Title')

# Left-connection to DisaggregatingDimension2ValueCode 
mort_300_filtered = mort_300_filtered.alias('mort').join(
    child_cause.alias('cause'),
    col('mort.DisaggregatingDimension2ValueCode') == col('cause.Code'),
    how='left'
).select(
    col('mort.*'),
    col('cause.Title').alias('Title')
)

# Delete unneeded columns (if any) 
mort_300_filtered = mort_300_filtered.drop('Code')

# Display of merged data 
mort_300_filtered.show(truncate=False)
```

```python
# Remove unneeded columns DisaggregatingDimension3 and DisaggregatingDimension3ValueCode
mort_300_filtered = mort_300_filtered.drop('DisaggregatingDimension3', 'DisaggregatingDimension3ValueCode')
```

```python
# Renaming columns 
mort_300_filtered = mort_300_filtered.withColumnRenamed('DisaggregatingDimension1', 'AgeGroup') \
    .withColumnRenamed('DisaggregatingDimension1ValueCode', 'AgeGroupValue') \
    .withColumnRenamed('DisaggregatingDimension2', 'ChildCause') \
    .withColumnRenamed('DisaggregatingDimension2ValueCode', 'ChildCauseCode')
# Show renamed data 
mort_300_filtered.show(truncate=False)
```

```python
# Delete unneeded columns ----step4-------------------
columns_to_drop = ['IndicatorCode', 'TimeDimension', 'AgeGroup', 'AgeGroupValue', 'ChildCause', 'ChildCauseCode', 'ParentLocationCode']
mort_300_filtered = mort_300_filtered.drop(*columns_to_drop)

# Renaming columns 
mort_300_filtered = mort_300_filtered.withColumnRenamed('Title', 'CausesOfDeath')

# Show renamed data 
mort_300_filtered.show(truncate=False)
```

```python
# Create a new column 'CountryOrMgheregOrParentLocation' and merge the SpatialDimensionValueCode with the value of ParentLocation
mort_300_filtered = mort_300_filtered.withColumn(
    'CountryOrMgheregOrParentLocation',
    concat_ws('', coalesce(col('SpatialDimensionValueCode'), lit('')), coalesce(col('ParentLocation'), lit('')))
)

# Delete the original two columns 
mort_300_filtered = mort_300_filtered.drop('SpatialDimensionValueCode', 'ParentLocation')

# Display of merged data 
mort_300_filtered.show(truncate=False)
```

```python
# after iteration-TimeDim
# Third model-TimeDim
# Step 1: Prepare Data
mort_300_filtered = mort_300_filtered.withColumn("TimeDim", mort_300_filtered["TimeDim"].cast(IntegerType()))

# Split the dataset into training set and test set
train_data_time, test_data_time = mort_300_filtered.randomSplit([0.8, 0.2], seed=1234)

# Step 2: Prepare Training Data
assembler_time = VectorAssembler(inputCols=['TimeDim'], outputCol='TimeDimVec')
train_assembled_time = assembler_time.transform(train_data_time).select('TimeDim', 'TimeDimVec', 'Log10_NumericValue')
test_assembled_time = assembler_time.transform(test_data_time).select('TimeDim', 'TimeDimVec', 'Log10_NumericValue')

# Apply Polynomial Expansion
poly_expansion = PolynomialExpansion(degree=2, inputCol="TimeDimVec", outputCol="polyFeatures")
train_poly = poly_expansion.transform(train_assembled_time)
test_poly = poly_expansion.transform(test_assembled_time)

# Step 3: Train the Linear Regression Model with polynomial features
lr_time = LinearRegression(featuresCol='polyFeatures', labelCol='Log10_NumericValue')
lr_time_model = lr_time.fit(train_poly)

# Step 4: Make predictions
train_predictions_time = lr_time_model.transform(train_poly)
test_predictions_time = lr_time_model.transform(test_poly)

# Step 5: Calculate Model Performance
evaluator_time = RegressionEvaluator(labelCol='Log10_NumericValue', predictionCol='prediction', metricName='rmse')
train_rmse_time = evaluator_time.evaluate(train_predictions_time)
train_r2_time = evaluator_time.evaluate(train_predictions_time, {evaluator_time.metricName: "r2"})
test_rmse_time = evaluator_time.evaluate(test_predictions_time)
test_r2_time = evaluator_time.evaluate(test_predictions_time, {evaluator_time.metricName: "r2"})

print(f"Train Root Mean Squared Error (RMSE): {train_rmse_time}")
print(f"Train R^2 Score: {train_r2_time}")
print(f"Test Root Mean Squared Error (RMSE): {test_rmse_time}")
print(f"Test R^2 Score: {test_r2_time}")


# Step 6: Plot Predicted Trend
# Convert predictions to Pandas DataFrame for plotting
train_predictions_df = train_predictions_time.select("TimeDim", "Log10_NumericValue", "prediction").toPandas()
test_predictions_df = test_predictions_time.select("TimeDim", "Log10_NumericValue", "prediction").toPandas()

# Combine train and test predictions for plotting
predictions_df = pd.concat([train_predictions_df, test_predictions_df])

# Plot the trend
plt.figure(figsize=(12, 8))
plt.scatter(predictions_df["TimeDim"], predictions_df["Log10_NumericValue"], color='black', label='Actual data')
plt.scatter(predictions_df["TimeDim"], predictions_df["prediction"], color='red', label='Predicted trend')
plt.xlabel('Time Dimension (Year)')
plt.ylabel('Log10 Numeric Value')
plt.title('Trend of Neonatal Mortality Over Time with Polynomial Regression')
plt.legend()

# Adjust x-axis to show integer years
plt.xticks(predictions_df["TimeDim"].unique(), rotation=45)
plt.gca().set_xticklabels(predictions_df["TimeDim"].unique().astype(int))

plt.show()
```
